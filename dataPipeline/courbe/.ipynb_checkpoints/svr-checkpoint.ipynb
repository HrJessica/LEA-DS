```python
import numpy as np
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
import json
import pickle
#importer le json et xsens
xsens_path = '../Xsens_profil.p'
def load_json(i):
    file_path = './json/Sujet'+i+'_1p_success.json'
    with open(file_path) as fl:
        sujet = json.load(fl)
    return sujet['frames']
def load_xsens(angleName,i):
    with open(xsens_path,'rb') as file:
        xsens_data = pickle.load(file)
    return xsens_data['Sujet'+i+'_T1'][angleName].tolist()

def get_data(angleName,i):
    list_data = []
    dict_=load_json(i)
    for key,values in dict_.items():
        list_data.append(values['angles'][angleName])
    return list_data 

    
# Generate sample data

n = ['1','5','6','7','9','10','12','13','14','15','18','21','22']
X=[]
y=[]
for k in n:
    X.append(get_data('Ft',k))
    y.append(load_xsens('Ft',k))
X = sum(X,[])
y= sum(y,[])
X = np.array(X)
y = np.array(y)
#standardisation
X =X.reshape(-1,1)
scaler = preprocessing.StandardScaler().fit(X)
X = scaler.transform(X)
#lissage par Savitzky-Golay ou savgol
np.set_printoptions(precision=2)
X = savgol_filter(X, 45, 2, mode='nearest')

# Fit regression model
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.1)

fig, axes = plt.subplots(nrows=1, figsize=(8, 8))
axes.scatter(X_train,y_train,s = 5,color = 'blue')
svr= SVR().fit(X_train,y_train)
yfit = svr.predict(X_test)
axes.scatter(X_test,yfit, s = 2 , color = 'red')
plt.show()
#save model
filename = 'model.pkl'
pickle.dump(svr, open(filename, 'wb'))
score = svr.score(X_train,y_train)
print('R_squares: ',score)
print('MSE: ',mean_squared_error(y_test,yfit))
fig, axes = plt.subplots(nrows=1, figsize=(25, 8))
axes.plot(range(len(X_test)),X_test,color = 'blue')
axes.plot(range(len(y_test)),y_test,color = 'red')
axes.grid()    
plt.show()
fig, axes = plt.subplots(nrows=1, figsize=(25, 8))
axes.plot(range(len(X_test)),X_test,color = 'blue')
axes.plot(range(len(yfit)),yfit,color = 'red')
axes.grid()    
plt.show()
```

```python
import pickle
import numpy as np
import json
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
xsens_path = '../Xsens_profil.p'
def load_json(i):
    file_path = './json/Sujet'+i+'_1p_success.json'
    with open(file_path) as fl:
        sujet = json.load(fl)
    return sujet['frames']
def load_xsens(angleName,i):
    with open(xsens_path,'rb') as file:
        xsens_data = pickle.load(file)
    return xsens_data['Sujet'+i+'_T1'][angleName].tolist()

def get_data(angleName,i):
    list_data = []
    dict_=load_json(i)
    for key,values in dict_.items():
        list_data.append(values['angles'][angleName])
    return list_data
# Generate sample data

n = ['4','11']
X=[]
y=[]
for k in n:
    X.append(get_data('Ft',k))
    y.append(load_xsens('Ft',k))

X = sum(X,[])
X = np.array(X)
y= sum(y,[])
X = X.reshape(-1,1)
scaler = preprocessing.StandardScaler().fit(X)
X = scaler.transform(X)

y = np.array(y)
y = np.array(y)
# load the model from disk
filename = 'model.pkl'
loaded_model = pickle.load(open(filename, 'rb'))
yfit = loaded_model.predict(X)

#tracer

fig, axes = plt.subplots(nrows=1, figsize=(25, 8))
axes.plot(range(len(yfit)),yfit,color = 'blue')
axes.plot(range(len(y)),y,color = 'red')
axes.grid()    
plt.show()

result = loaded_model.score(X,y)
print(result)
```
